### Conduct your test. (5 points)
Explain (1) the statistical significance of your result, and (2) the practical significance of your result.  Make sure you relate your findings to the original research question.

````r
t.test(df$votersurvey.ftpolice, df$votersurvey.ftjournal, paired=TRUE)
p_ttest_vals <- t.test(df$votersurvey.ftpolice, df$votersurvey.ftjournal, paired=TRUE)

pval <- p_ttest_vals$p.value
meanval <- p_ttest_vals$estimate
stddevval <- p_ttest_vals$stderr    
dfval <- p_ttest_vals$parameter    


print("PRINTING VALUES FROM T-TEST ---")
mean_ftpolice <- mean(df$votersurvey.ftpolice)
mean_ftjournal <- mean(df$votersurvey.ftjournal)
mean_ftpolice
mean_ftjournal
pval
as.numeric(meanval)
stddevval
dfval
cohensd <- as.numeric(meanval) / stddevval/as.numeric(dfval)
cohensd

print("PRINTING RESULT INFERENCES FROM T-TEST ---")

if(pval < 0.05 && (mean_ftpolice < mean_ftjournal)){ 
	print("statistically significant difference")
	print("Voters have more respect for journalists")
} else if(pval < 0.05 && (mean_ftpolice > mean_ftjournal)){ 
	print("statistically significant difference")
	print("Voters have more respect for police")
} else { 
	print("No statistically significant difference")
}
    
````

#### STATISTICAL SIGNIFICANCE: 

p-value very low; paired t-test is highly statistically significant. Therefore, means of both results are different. 

#### PRACTICAL SIGNIFICANCE: 

Cohen's d is very small ~0.00549104410943403. Which means that the effect size is not significant
